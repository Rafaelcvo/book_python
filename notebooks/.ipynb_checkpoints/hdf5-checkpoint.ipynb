# Usando o formato HDF5

```python
import numpy as np
import pandas as pd
```

```python
frame = pd.DataFrame({'a':np.random.randn(100)})
```

```python
store = pd.HDFStore('mydata.h5')
```

```python
store['obj1'] = frame
```

```python
store['obj1_col'] = frame['a']
```
